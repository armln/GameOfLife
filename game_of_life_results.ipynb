```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import multiprocessing as mp
```

```python
with open('results.txt','r') as f:
    T=[x.strip().split(' ') for x in f][0]

TT = []
for x in T:
    TT.append(float(x))
```

```python
TT
```

```python
van = np.array(TT[:6])
van
```

```python
tr_2 = TT[6:12]
tr_2
```

```python
tr_3 = TT[12:18]
tr_3
```

```python
tr_4 = TT[18:24]
tr_4
```

```python
plt.figure(figsize=(16,9))
plt.plot([50,100,150,200,250,300],van,'o-', label = 'No threading')
plt.plot([50,100,150,200,250,300],tr_2,'o-', label = '2 threads')
plt.plot([50,100,150,200,250,300],tr_3,'o-', label = '3 threads')
plt.plot([50,100,150,200,250,300],tr_4,'o-', label = '4 threads')
plt.legend()
plt.show()
```
